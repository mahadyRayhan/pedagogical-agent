```python
from transformers import BertTokenizer, BertForMaskedLM, Trainer, TrainingArguments
from torch.utils.data import Dataset
import torch

# Custom Dataset class
class TextDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __len__(self):
        return len(self.encodings['input_ids'])

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = item['input_ids']
        return {key: torch.tensor(val) for key, val in item.items()}

combined_text = ""
for doc in all_documents:
    combined_text += doc.text

# Load MatSciBERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("m3rg-iitd/matscibert")
model = BertForMaskedLM.from_pretrained("m3rg-iitd/matscibert")

# Tokenize the combined text
inputs = tokenizer([combined_text], return_tensors="pt", truncation=True, padding=True, max_length=512)

# Prepare the dataset using the custom TextDataset class
train_dataset = TextDataset(inputs)

# Define training arguments
training_args = TrainingArguments(
    output_dir="/content/matscibert_finetuned",  # This is where the model will be saved
    per_device_train_batch_size=1,
    num_train_epochs=3,
    save_steps=10_000,
    save_total_limit=2,
    logging_dir="/content/logs",  # Logs will be saved here
    logging_steps=500,
    gradient_accumulation_steps=8,
    fp16=True,  # Enable mixed precision for faster training
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)

trainer.train()

trainer.save_model("/content/matscibert_finetuned")
tokenizer.save_pretrained("/content/matscibert_finetuned")

```

```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch

# Load MatSciBERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("m3rg-iitd/matscibert")
model = BertForQuestionAnswering.from_pretrained("m3rg-iitd/matscibert")

# Function to answer questions based on the model's knowledge
def answer_question(question):
    # Tokenize the question
    inputs = tokenizer(question, return_tensors='pt')

    # Forward pass through the model
    outputs = model(**inputs)

    # Get the predicted answer span
    start_logits, end_logits = outputs.start_logits, outputs.end_logits
    start_index = torch.argmax(start_logits)
    end_index = torch.argmax(end_logits)

    # Decode the answer
    answer_tokens = inputs['input_ids'][0][start_index:end_index + 1]
    answer = tokenizer.decode(answer_tokens)

    return answer.strip()

# Example question
question = "What is a good temperature to grow CNTs?"
answer = answer_question(question)

print("Question:", question)
print("Predicted Answer:", answer)
```

```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch

# Load the fine-tuned MatsciBERT model and tokenizer
model = BertForQuestionAnswering.from_pretrained("/content/matscibert_finetuned")
tokenizer = BertTokenizer.from_pretrained("/content/matscibert_finetuned")

def answer_question(question, context):
    # Encode the inputs with truncation and padding
    inputs = tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt", truncation=True, padding=True, max_length=512)

    # Get the input IDs and attention masks
    input_ids = inputs['input_ids']
    attention_mask = inputs['attention_mask']

    # Get model predictions
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)

    # Get the start and end scores
    start_scores, end_scores = outputs.start_logits, outputs.end_logits

    # Get the most likely start and end indices
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores) + 1  # +1 because the end index is inclusive

    # Ensure the indices are valid and within the input_ids length
    if start_index < end_index and end_index <= input_ids.size(1):
        answer_ids = input_ids[0][start_index:end_index]
        answer = tokenizer.decode(answer_ids, skip_special_tokens=True)
    else:
        answer = "No valid answer found."

    return answer

# Using the function
context = combined_text[:1000]  # Example of truncating the context to the first 1000 characters
question = "What are the recent advancements in carbon nanotube research?"

# Get the answer
answer = answer_question(question, context)
print("Predicted Answer:", answer)
```

```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch

# Load MatSciBERT tokenizer and model
mat_tokenizer = BertTokenizer.from_pretrained("/content/matscibert_finetuned")
mat_model = BertForQuestionAnswering.from_pretrained("/content/matscibert_finetuned")

def mat_answer_question(question, context):
    inputs = mat_tokenizer.encode_plus(question, context, return_tensors="pt")
    with torch.no_grad():
        outputs = mat_model(**inputs)
        start_scores = outputs.start_logits
        end_scores = outputs.end_logits

    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores) + 1
    answer = mat_tokenizer.convert_tokens_to_string(mat_tokenizer.convert_ids_to_tokens(inputs['input_ids'][0][start_index:end_index]))

    return answer.strip()

context = (
    "Carbon nanotubes (CNTs) are cylindrical structures made of carbon atoms arranged in a hexagonal lattice. "
    "They exhibit extraordinary mechanical strength, with tensile strengths exceeding those of steel, while being remarkably lightweight. "
    "Due to their unique electrical properties, CNTs can conduct electricity better than copper and are often used in various electronic components. "
    "There are two main types of carbon nanotubes: single-walled (SWCNTs) and multi-walled (MWCNTs). "
    "Synthesis methods include chemical vapor deposition (CVD), laser ablation, and arc discharge. "
    "In electronics, CNTs are utilized in transistors, sensors, and conductive films. "
    "In materials science, they enhance the mechanical properties of composites and are used in the development of lightweight, high-strength materials. "
    "In nanotechnology, CNTs serve as carriers for drug delivery and in the creation of nanoscale devices. "
    "Research continues to explore their applications in energy storage, such as in batteries and supercapacitors, and in environmental remediation."
)


# dataset with questions
questions = [
    "What are the key factors influencing the growth of carbon nanotubes?",

    "What are the most common synthesis methods used for carbon nanotubes in recent research?",

    "How does the choice of catalyst affect the quality and yield of carbon nanotubes in experiments?",

    "What are the effects of varying gas flow rates during CNT synthesis in recent literature?",

    "How are CNTs being used in energy storage devices, particularly in batteries and supercapacitors?",

    "What potential do CNTs have in biomedical applications, such as drug delivery or tissue engineering?",

    "How does the performance of CNTs compare to other nanomaterials like graphene in terms of electrical conductivity?",

    "What are the key differences in the experimental results of CNT growth between different research papers?",

    "What are the latest innovations in CNT-based sensors for environmental monitoring?",

    "How has the use of machine learning or AI influenced CNT research, particularly in optimizing synthesis parameters?"
]

# Generate answers using MatSciBERT
answers = [mat_answer_question(q, context) for q in questions]

# Prepare dataset for fine-tuning T5
fine_tune_data = list(zip(questions, answers))

for question, answer in zip(questions, answers):
    print(f"Question: {question}\nAnswer: {answer}\n")
```

```python
from torch.utils.data import Dataset

class T5Dataset(Dataset):
    def __init__(self, data, tokenizer):
        self.tokenizer = tokenizer
        self.data = data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        question, answer = self.data[idx]
        input_encodings = self.tokenizer(f"question: {question} context: ", return_tensors='pt', padding="max_length", max_length=512, truncation=True)
        target_encodings = self.tokenizer(answer, return_tensors='pt', padding="max_length", max_length=100, truncation=True)

        input_ids = input_encodings['input_ids'].flatten()
        attention_mask = input_encodings['attention_mask'].flatten()
        labels = target_encodings['input_ids'].flatten()

        return {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'labels': labels
        }

# Create the dataset
t5_dataset = T5Dataset(fine_tune_data, tokenizer)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments

# Load T5 tokenizer and model
tokenizer = T5Tokenizer.from_pretrained("t5-small")  # You can choose a larger version
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Define training arguments
training_args = TrainingArguments(
    output_dir="./t5_finetuned_matscibert",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    save_steps=10_000,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=500,
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=t5_dataset,
)

# Start training
trainer.train()

# Save the fine-tuned model and tokenizer
trainer.save_model("./t5_finetuned_matscibert_new")
tokenizer.save_pretrained("./t5_finetuned_matscibert_new")
```

```python
import torch

def ask_question(question, context):
    # Encode the input question and context
    input_text = f"question: {question} context: {context}"
    input_encodings = tokenizer(input_text, return_tensors='pt', padding=True, truncation=True, max_length=512)

    # Move input tensors to the same device as the model (CPU/GPU)
    input_ids = input_encodings['input_ids'].to(model.device)
    attention_mask = input_encodings['attention_mask'].to(model.device)

    # Generate the answer
    with torch.no_grad():
        output_ids = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_length=100)

    # Decode the generated answer
    answer = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return answer.strip()
```
